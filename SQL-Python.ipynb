```python
import pandas as pd
import sqlite3
```

```python
#read the CSV
#connect to a database
def load_data(data, table_name):
    df = pd.read_csv(data)
    conn = sqlite3.connect("Database.db") #if the db does not exist, this creates a Any_Database_Name.db file in the current directory
    #store your table in the database:
    df.to_sql(table_name, conn)
    #read a SQL Query out of your database and into a pandas dataframe
    sql_string = 'SELECT * FROM ' + table_name
    df = pd.read_sql(sql_string, conn)
    print(df)
```

```python
files = ["Census_Data_-_Selected_socioeconomic_indicators_in_Chicago__2008___2012-v2.csv",
        "Chicago_Public_Schools_-_Progress_Report_Cards__2011-2012-v3.csv",
        "Chicago_Crime_Data-v2.csv"]
names = ["census", "school", "crime"]
```

```python
for f, n in zip(files, names):
    load_data(f, n)
```
